```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from pathlib import Path
import os
```

```python
keyword_collection_orals = {}
base_path = Path("processed_data")
for file in os.listdir(base_path):
    if file.endswith(".txt") and file.startswith("manual"):
        with open(base_path / file, "r") as f:
            lines = ",".join(f.readlines())
            conf_name = file.split("_")[3].split(".")[0]
            conf_year = file.split("_")[4].split(".")[0]
            keyword_collection_orals[f"{conf_name}_{conf_year}"] = lines
```

```python
for k in keyword_collection_orals.keys():
    plt.figure()
    wc = WordCloud(background_color='white', colormap = 'bone').generate(keyword_collection_orals[k])
    plt.title(f"Keywords of Orals at {k}")
    plt.imshow(wc)
    plt.axis("off")
```
